```python
import numpy as np
import matplotlib.pyplot as plt
import skimage
from skimage.color import rgb2gray
import geomstats.geometry.discrete_curves
from geomstats.geometry.euclidean import Euclidean
from geomstats.geometry.discrete_curves import DiscreteCurves

from tqdm.notebook import tqdm
```

```python
from skimage.filters import sobel
elevation_map = sobel(b)
```

```python
markers = np.zeros_like(b)
markers[b < 0.2] = 1
markers[b > .8] = 2
```

```python
from skimage import measure

# Find contours at a constant value of 0.8
contours = measure.find_contours(segmentation, 1)


# Display the image and plot all contours found
fig, ax = plt.subplots()
ax.imshow(segmentation, cmap=plt.cm.gray)

for contour in contours:
    ax.plot(contour[:, 1], contour[:, 0], linewidth=2)

ax.axis('image')
ax.set_xticks([])
ax.set_yticks([])
plt.show()
```

```python
import geomstats.backend as gs

def interpolate(curve, nb_points):
    """Interpolate a discrete curve with nb_points from a discrete curve.

    Returns
    -------
    interpolation : discrete curve with nb_points points
    """
    old_length = curve.shape[0]
    interpolation = gs.zeros((nb_points, 2))
    incr = old_length / nb_points
    pos = 0
    for i in range(nb_points):
        index = int(gs.floor(pos))
        interpolation[i] = curve[index] + (pos - index) * (
            curve[(index + 1) % old_length] - curve[index]
        )
        pos += incr
    return interpolation


k_sampling_points = 1000

def preprocess(curve, tol=1e-10):
    """Preprocess curve to ensure that there are no consecutive duplicate points.

    Returns
    -------
    curve : discrete curve
    """

    dist = curve[1:] - curve[:-1]
    dist_norm = np.sqrt(np.sum(np.square(dist), axis=1))

    if np.any( dist_norm < tol ):
        for i in range(len(curve)-2):
            if np.sqrt(np.sum(np.square(curve[i+1] - curve[i]), axis=0)) < tol:
                curve[i+1] = (curve[i] + curve[i+2]) / 2

    return curve

def center(point):

    mean = gs.mean(point, axis=-2)
    
    return point - mean[..., None, :]
```

```python
A1 = contours[6]
A2 = contours[11]
```

```python
B2 = interpolate(A2,200)
B1 = interpolate(A1,200)
```

```python
plt.plot(B2[:,0],B2[:,1])
```

```python
from geomstats.geometry.pre_shape import PreShapeSpace

M_AMBIENT = 2

PRESHAPE_SPACE = PreShapeSpace(m_ambient=M_AMBIENT, k_landmarks=200)
PRESHAPE_METRIC = PRESHAPE_SPACE.embedding_space.metric

curve = PRESHAPE_SPACE.projection(B1)

base_curve = PRESHAPE_SPACE.projection(B2)

nb_sampling = len(curve)

distances = gs.zeros(nb_sampling)

base_curve = gs.array(base_curve)

for shift in range(nb_sampling):
    reparametrized = [curve[(i + shift) % nb_sampling] for i in range(nb_sampling)]
    aligned = PRESHAPE_SPACE.align(point=gs.array(reparametrized), base_point=base_curve)
    distances[shift] = PRESHAPE_METRIC.norm(gs.array(aligned) - gs.array(base_curve))
    
shift_min = gs.argmin(distances)

reparametrized_min = [curve[(i + shift_min) % nb_sampling] for i in range(nb_sampling)]

aligned_curve = PRESHAPE_SPACE.align(point=gs.array(reparametrized_min), base_point=base_curve)
```

```python
C1 = center(aligned_curve)
C2 = center(base_curve)
```

```python
plt.plot(C1[:,0],C1[:,1])
plt.plot(C2[:,0],C2[:,1])
```

```python
n_times = 10
times = gs.linspace(0.0, 1.0, n_times)

r2 = Euclidean(dim=2)

metric = geomstats.geometry.discrete_curves.SRVMetric(r2)

hgeod_fun = metric.geodesic(preprocess(C1), preprocess(C2))
hgeod = hgeod_fun(times)
```

```python
plt.figure(figsize=(7, 7))
plt.plot(hgeod[0, :, 0], hgeod[0, :, 1], "b")
for i in range(1, n_times - 1):
    plt.plot(hgeod[i, :, 0], hgeod[i, :, 1], "k")
plt.plot(hgeod[-1, :, 0], hgeod[-1, :, 1], "r")
plt.title("Horizontal geodesic")
plt.show()
```

```python
count = 0

good_ind = []

good_a = []

for i in contours:
    
    if len(i) < 350:
        plt.plot(i[:,0],i[:,1])
        good_ind.append(count)
        good_a.append(i)
    
    count += 1
```

```python
f = [0,3,5,6,13,17,18]

a_good = []

for i in range(18):
    if i not in f:
        a_good.append(good_a[i])
```

```python
dist = np.zeros((12,12))

PRESHAPE_SPACE = PreShapeSpace(m_ambient=M_AMBIENT, k_landmarks=100)
PRESHAPE_METRIC = PRESHAPE_SPACE.embedding_space.metric

Curves = []

for i in tqdm(range(12)):
    
    for j in tqdm(range(12)):
        
        B2 = interpolate(a_good[i],100)
        B1 = interpolate(a_good[j],100)
        
        curve = preprocess(B1)

        base_curve = preprocess(B2)

        nb_sampling = len(curve)

        distances = gs.zeros(nb_sampling)

        base_curve = gs.array(base_curve)

        for shift in range(nb_sampling):
            reparametrized = [curve[(i + shift) % nb_sampling] for i in range(nb_sampling)]
            aligned = PRESHAPE_SPACE.align(point=gs.array(reparametrized), base_point=base_curve)
            distances[shift] = PRESHAPE_METRIC.norm(gs.array(aligned) - gs.array(base_curve))

        shift_min = gs.argmin(distances)

        reparametrized_min = [curve[(i + shift_min) % nb_sampling] for i in range(nb_sampling)]

        aligned_curve = PRESHAPE_SPACE.align(point=gs.array(reparametrized_min), base_point=base_curve)
        
        C1 = center(aligned_curve)
        
        C2 = center(base_curve)
        
        hgeod_fun = metric.geodesic(preprocess(C1), preprocess(C2))

        dist[i,j] = metric.dist(C1,C2)
        
    Curves.append(C2)
```

```python
from sklearn.decomposition import PCA

pca = PCA()

data = metric.log(Curves,base_point = Curves[-1])

Euc_stack = []

for i in data:
    np.hstack(i).shape
    Euc_stack.append(np.hstack(i))
    
X = pca.fit_transform(Euc_stack)

plt.scatter(X[:,1],X[:,2])
```

```python
n_times = 12

count = 1

fig = plt.figure(figsize=(10, 2))

curves = []

for i in np.argsort(X[:,1]):
    fig.add_subplot(2, 12,count)
    B1 = interpolate(a_good[i],100)

    curve = PRESHAPE_SPACE.projection(preprocess(B1))
    
    curves.append(curve)

    plt.plot(B1[:, 0], B1[:, 1])
    plt.axis("equal")
    plt.axis("off")
    count += 1
```
